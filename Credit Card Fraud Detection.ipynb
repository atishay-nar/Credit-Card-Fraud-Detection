```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
# load csv
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
# describing the data
print(data.shape)
data.describe()
```

```python
# fradulent cases in the dataset
fraud = data[data["Class"] == 1]
valid = data[data["Class"] == 0]
fraudPercentage = len(fraud) / float(len(valid))
print(f"percentage of fradulent transactions: {fraudPercentage*100}%")
print(f"number of fradulent transactions: {len(fraud)}")
print(f"number of valid transactions: {len(valid)}")
```

```python
# fraud amount details
fraud.Amount.describe()
```

```python
# valid amount details
valid.Amount.describe()
```

```python
# plot the correlation matrix to see which features are relevant

corrmat = data.corr()
fig = plt.figure(figsize=(10, 7))
sns.heatmap(corrmat, vmax=0.8, square=True)
```

```python
# seperate data into X and Y components
X = data.drop(["Class"], axis=1)
Y = data["Class"]
print(X.shape)
print(Y.shape)

# get values only from columns
xData = X.values
yData = Y.values
```

```python
# split the data into test-train sets
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(
    xData, yData, train_size=0.2, random_state=24
)
```

```python
# random forest model
from sklearn.ensemble import RandomForestClassifier

# create rfc with the x and y training sets
rfc = RandomForestClassifier()
rfc.fit(xTrain, yTrain)

# predict using x test
yPred = rfc.predict(xTest)
```

```python
# evaluating the classifier
from sklearn.metrics import (
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    matthews_corrcoef,
    confusion_matrix,
)

n_outliers = len(fraud)
n_errors = (yTest != yPred).sum()

acc = accuracy_score(yTest, yPred)
pre = precision_score(yTest, yPred)
rec = recall_score(yTest, yPred)
f1 = f1_score(yTest, yPred)
mcc = matthews_corrcoef(yTest, yPred)

print(f"The accuracy score is {acc}")
print(f"The precision score is {pre}")
print(f"The recall score is {rec}")
print(f"The f1 score is {f1}")
print(f"The mathhews correlation coefficent is {mcc}")
```

```python
# making a confusion matrix
labels = ["Valid", "Fraudulent"]
confMatrix = confusion_matrix(yTest, yPred)
plt.figure(figsize=(6, 6))
sns.heatmap(confMatrix, xticklabels=labels, yticklabels=labels, annot=True, fmt="d")
plt.title("Confusion Matrix")
plt.ylabel("True Class")
plt.xlabel("Predicted Class")
```
